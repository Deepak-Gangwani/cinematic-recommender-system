```python
# I will convert these string in the above data into integer for fetch purpose using one library
import ast
```

```python
# creating a function to convert the above data into single array which value of name key
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
# Implementing the function on dataset to seperate the generes data into list on all the rows
movies["genres"]=movies["genres"].apply(convert)
```

```python
movies["genres"]
```

```python
movies.head(1)
```

```python
# The same process will be done to keywords column using the same above steps
# Implementing the function on dataset to seperate the keywords data into list on all the rows
movies["keywords"]=movies["keywords"].apply(convert)
movies["genres"]
```

```python
movies.head(1)
```

```python
# creating a function to convert the above data into single array which value of name key
# In cast column I will fetch those actor which are in the first top 3 that I will show
def convert3(obj):
    l=[]
    count=0
    for i in ast.literal_eval(obj):
        if(count<3):
            l.append(i["name"])
            count=count+1
        else:
            break
    return l
```

```python
# Selecting the cast column from dataset for filtering the top 3 actors in every movie 
movies["cast"]=movies["cast"].apply(convert3)
movies["cast"]
```

```python
movies.head(1)
```

```python
# now I will do filtering in crew column
# let just see the column before filtering
movies["crew"][0]
```

```python
# Now I will filter that data from crew column whose job role is director with the director name

# creating a function to convert the above data into single array which value of name key with director name
# In crew column I will fetch those director name who the directed the movie
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if (i["job"]=="Director"):
            l.append(i["name"])
            break
    return l
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
movies["crew"]
```

```python
movies.head(1)
```

```python
# Converting overview row data into list for after concatenation
movies["overview"][0]
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
movies["overview"]
```

```python
movies.head(1)
```

```python
# Now I have to convert the cast data by removing whitespace
# example
# befor:-Sam Worthington
# after:-SamWorthington
# I am doing this filtering for not recommending the same name of different movie 
# If their is one sam in cast and another sam in crew the machine will get confuse and this will recommend wring movies
# so we are joining the first name and last name by removing same to find the person easily
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])#list comprehension for filtering
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])#list comprehension for filtering
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])#list comprehension for filtering
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])#list comprehension for filtering
```

```python
movies.head(1)
```

```python
# Now I will create a tag column which will cancatenate all 4 list columns in one tags columns
movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
movies.head(1)
```

```python
# Now removing the list column because I have cancatenate them in one column that is tags
# creating a new data frame which will have 3 column that is title and tags for recommendation analysis
new_df=movies[["movie_id","title","tags"]]
new_df.head()
```

```python
#Now I will convert the the tags list into string by using list comprehension method
new_df["tags"]=new_df["tags"].apply(lambda x:" ".join(x))
new_df["tags"]
```

```python
new_df["tags"][0]# particular first row tags column converted into string
```

```python
new_df.head()
```

```python
# converting the tags data into lowercase due to some case-sensitive reason
new_df["tags"]=new_df["tags"].apply(lambda x:x.lower())
new_df.head()
```

# Text Vectorization Process Will Be Done After Filtering All the Dataset

```python
# Now I have to identify the similarity between the first two columns of tags using some concept
new_df["tags"][0]
```

```python
new_df["tags"][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_df["tags"]).toarray()
vectors
```

```python
# Applying steming method which convert the singular and plural words into one word
# example
# Before:-["loved","loving","love"]
# After:-["Love","love","love"]
!pip install nltk
```

```python
import nltk # it is a natural laguage processing library in python
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
# creating function with help of library which will convert the plural word into singular
# example
ps.stem("loving")
```

```python
ps.stem("loved")
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y) #converting list into string again after doing changes
```

```python
# applying stem function in dataframe
new_df["tags"]=new_df["tags"].apply(stem)
new_df["tags"]
```

```python
# After using steming let see the changes in 5000 words
cv.get_feature_names()
```

# Creating a method to calculate the cosine distance between every movie which get compared

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)#this will check the 4806 times similarity between every single movie
similarity 
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
# Creating a function which which will return the top 5 movies which nearest similar to the given data
def recommend(movie):
    movie_index = new_df[new_df["title"]==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
#         print(i[0])
        print(new_df.iloc[i[0]].title)
        
    
```

```python
recommend("Avatar")
```

```python
recommend("Batman Begins")
```
